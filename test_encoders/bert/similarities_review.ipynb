```python
def review_similarities_file(fileName):
    f = open('similarities/' + fileName, 'r')

    resume_similarities = [line.split() for line in f.readlines()]

    # print(resume_similarities)

    sorted_resume_similarities = sorted(resume_similarities, key=lambda resume_similarities: resume_similarities[3], reverse=True)

    for resume in sorted_resume_similarities:
        print(str(resume[0]).ljust(6, ' ') + ' ' + str(resume[1]).ljust(10, ' ') + ' ' + str(resume[3]).ljust(20, ' ') + ' ' + str(resume[2]))
```

```python
review_similarities_file('s2s_slavepe_avecs.txt')
```

```python
review_similarities_file('d2d_dlavepe_avecs.txt')
```
